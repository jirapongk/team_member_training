## Merging and Joining DataFrames
- Concatenation: concat()
    - Concatenation is used to combine pandas objects along a particular axis. 
- Merging: merge()
    - Merging is used to combine DataFrames based on common columns or indices.

```python
# Create two DataFrames
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                    'B': ['B0', 'B1', 'B2']})
df2 = pd.DataFrame({'A': ['A3', 'A4', 'A5'],
                    'B': ['B3', 'B4', 'B5']})

# Concatenate along rows (axis=0)
result = pd.concat([df1, df2])

print(result)
```

```python
# Create two DataFrames
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})

# Merge DataFrames on the 'key' column
result = pd.merge(left, right, on='key')

print(result)
```

## Time Series Data Handling
- Working with dates and times
- Resampling : Resampling allows you to change the frequency of your time series data
- Rolling Windows : Rolling windows are used for moving calculations
- Lead and lag : operations shift the data forward or backward

```python
# Create a date range
date_range = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')

# Create a DataFrame with dates and random values
df = pd.DataFrame({
    'date': date_range,
    'value': np.random.randn(len(date_range))
})

# Set the date column as the index
df.set_index('date', inplace=True)

print(df.head())
```

```python
# Accessing time-related properties
print("\nYear:")
print(df.index.year)
print("\nMonth:")
print(df.index.month)
print("\nWeekday:")
print(df.index.weekday)
```

```python
# Filtering data for a specific date range
print("\nData for first 5 days:")
print(df['2023-01-01':'2023-01-05'])
```

```python
# Resample to monthly frequency
monthly_mean = df.resample('M').mean()
print("Monthly mean:")
print(monthly_mean.head())
```

```python
# Resample to quarterly frequency
quarterly_sum = df.resample('Q').sum()
print("\nQuarterly sum:")
print(quarterly_sum.head())
```

```python
# Compute 3-day rolling mean
df['3_day_rolling_mean'] = df['value'].rolling(window=3).mean()

print(df.head(10))
```

```python
# Create a lag of 1 day
df['lag_1'] = df['value'].shift(1)

# Create a lead of 1 day
df['lead_1'] = df['value'].shift(-1)

print(df.head())
```
